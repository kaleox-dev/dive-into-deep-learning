```python
import torch
```

```python
x = torch.arange(4.0)
x
```

```python
# Can also create x = torch.arange(4.0, requires_grad=True)
x.requires_grad_(True)
x.grad  # The gradient is None by default
```

```python
y = 2 * torch.dot(x, x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad == 4 * x
```

```python
x.grad.zero_() # reset the gradient
y = x.sum()
y.backward()
x.grad
```

```python
x.grad.zero_()
y = x * x
y.backward(gradient=torch.ones(len(y)))  # Faster: y.sum().backward()
x.grad
```

```python
x.grad.zero_()
y = x * x
u = y.detach()
z = u * x

z.sum().backward()
x.grad == u
```

```python
x.grad.zero_()
y.sum().backward()
x.grad == 2 * x
```
